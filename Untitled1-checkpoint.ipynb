```python
import os
import requests
import zipfile

# Create dataset directory
dataset_path = "D:/coco_dataset" 
os.makedirs(dataset_path, exist_ok=True)

# MS COCO URLs
images_url = "http://images.cocodataset.org/zips/train2017.zip"
annotations_url = "http://images.cocodataset.org/annotations/annotations_trainval2017.zip"

# Download function with file size check
def download_and_extract(url, save_path):
    filename = url.split("/")[-1]
    file_path = os.path.join(save_path, filename)

    # Download if not exists or file is too small
    if not os.path.exists(file_path) or os.path.getsize(file_path) < 1024 * 1024:  # Check if file is smaller than 1MB
        print(f"Downloading {filename}...")
        response = requests.get(url, stream=True)
        with open(file_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    f.write(chunk)
        print(f"Downloaded {filename}.")

    # Check if file is valid before extracting
    try:
        with zipfile.ZipFile(file_path, "r") as zip_ref:
            zip_ref.extractall(save_path)
            print(f"Extracted {filename}.")
    except zipfile.BadZipFile:
        print(f"Error: {filename} is corrupted. Try deleting and re-running the script.")

# Download images and annotations
download_and_extract(images_url, dataset_path)
download_and_extract(annotations_url, dataset_path)

print("MS COCO dataset downloaded and extracted successfully!")
```

# Data Preprocessing Code


```python
 import os
import json
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import Sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from pycocotools.coco import COCO

# Step 1: Define Paths
dataset_dir = "D:/coco_dataset/train2017"
annotations_path = "D:/coco_dataset/annotations/instances_train2017.json"

# Step 2: Define Animal Categories from COCO Dataset
animal_categories = {
    17: "cat", 18: "dog", 19: "horse", 20: "sheep", 21: "cow",
    22: "elephant", 23: "bear", 24: "zebra", 25: "giraffe"
}

# Load COCO annotations
coco = COCO(annotations_path)

# Step 3: Get Image IDs for Selected Animal Categories
animal_image_ids = []
image_id_to_label = {}

for cat_id, name in animal_categories.items():
    img_ids = coco.getImgIds(catIds=[cat_id])
    for img_id in img_ids:
        image_id_to_label[img_id] = list(animal_categories.keys()).index(cat_id)  # Assign numeric labels
    animal_image_ids.extend(img_ids)

print(f"Total animal images found: {len(animal_image_ids)}")

# Step 4: Train-Test Split (80% Train, 20% Test)
train_ids, test_ids = train_test_split(animal_image_ids, test_size=0.2, random_state=42)

# Step 5: Create Data Generator Class
class CocoDataGenerator(Sequence):
    def __init__(self, image_ids, batch_size=32, img_size=(128, 128), shuffle=True):
        self.image_ids = image_ids
        self.batch_size = batch_size
        self.img_size = img_size
        self.shuffle = shuffle
        self.on_epoch_end()
    
    def __len__(self):
        return int(np.floor(len(self.image_ids) / self.batch_size))
    
    def __getitem__(self, index):
        batch_ids = self.image_ids[index * self.batch_size:(index + 1) * self.batch_size]
        X, Y = self.__data_generation(batch_ids)
        return X, Y

    def __data_generation(self, batch_ids):
        X, Y = [], []
        for img_id in batch_ids:
            img_info = coco.loadImgs(img_id)[0]
            img_path = os.path.join(dataset_dir, img_info["file_name"])
            
            if os.path.exists(img_path):
                img = cv2.imread(img_path)
                img = cv2.resize(img, self.img_size)  # Resize images
                img = img / 255.0  # Normalize
                X.append(img)
                Y.append(image_id_to_label[img_id])  # Assign label
            else:
                print(f"Warning: Image not found at {img_path}")  # Log missing files
        
        return np.array(X), np.array(Y)

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.image_ids)

# Step 6: Create Train and Test Data Generators
batch_size = 32
train_generator = CocoDataGenerator(train_ids, batch_size=batch_size)
test_generator = CocoDataGenerator(test_ids, batch_size=batch_size, shuffle=False)

# Step 7: Define CNN Model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2, 2),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(animal_categories), activation='softmax')  # Output layer
])

# Step 8: Compile Model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Step 9: Train Model Using Generator
history = model.fit(train_generator, validation_data=test_generator, epochs=50)

# Step 10: Save Model and Weights Separately
model_path = "D:/animal_detection_model.h5"
weights_path = "D:/animal_detection_weights.h5"

model.save(model_path)  # Save full model
model.save_weights(weights_path)  # Save only weights

print(f"Model saved at: {model_path}")
print(f"Weights saved at: {weights_path}")

# Step 11: Print Final Accuracy (Convert to Percentage)
final_training_accuracy = history.history['accuracy'][-1] * 100  # Last training accuracy
final_validation_accuracy = history.history['val_accuracy'][-1] * 100  # Last validation accuracy

print(f"Final Training Accuracy: {final_training_accuracy:.2f}%")
print(f"Final Validation Accuracy: {final_validation_accuracy:.2f}%")
```

# GUI for Image & Video Detection (Tkinter & OpenCV)

```python
import os
import cv2
import numpy as np
import tensorflow as tf
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

# Load Trained Model
model_path = "D:/animal_detection_model.h5"
model = tf.keras.models.load_model(model_path)

# Define Animal Categories
animal_categories = {
    17: "cat", 18: "dog", 19: "horse", 20: "sheep", 21: "cow",
    22: "elephant", 23: "bear", 24: "zebra", 25: "giraffe"
}
class_labels = list(animal_categories.values())

# Define Carnivorous Animals for Warning System
carnivorous_animals = ["bear", "dog"]

# Function to Process Image & Predict
def process_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (128, 128))  # Resize
    img = img / 255.0  # Normalize
    img = np.expand_dims(img, axis=0)  # Expand dimensions for model input

    prediction = model.predict(img)
    predicted_class = np.argmax(prediction)
    predicted_label = class_labels[predicted_class]
    
    return predicted_label

# Function to Open File Dialog and Detect Animal
def upload_image():
    file_path = filedialog.askopenfilename()
    if file_path:
        predicted_label = process_image(file_path)
        
        # Display Image
        img = Image.open(file_path)
        img = img.resize((300, 300))
        img = ImageTk.PhotoImage(img)
        panel.config(image=img)
        panel.image = img
        
        # Update Label
        result_label.config(text=f"Detected: {predicted_label}", fg="green" if predicted_label not in carnivorous_animals else "red")

# Function for Live Webcam Detection
def live_detection():
    cap = cv2.VideoCapture(0)
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess frame
        img = cv2.resize(frame, (128, 128))
        img = img / 255.0
        img = np.expand_dims(img, axis=0)

        # Predict
        prediction = model.predict(img)
        predicted_class = np.argmax(prediction)
        predicted_label = class_labels[predicted_class]

        # Display Prediction on Frame
        color = (0, 255, 0) if predicted_label not in carnivorous_animals else (0, 0, 255)  # Red for carnivores
        cv2.putText(frame, predicted_label, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

        # Show Frame
        cv2.imshow("Live Animal Detection", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Create GUI Window
root = tk.Tk()
root.title("Animal Detection")

# Upload Button
upload_btn = tk.Button(root, text="Upload Image", command=upload_image, font=("Arial", 12))
upload_btn.pack()

# Panel to Show Image
panel = tk.Label(root)
panel.pack()

# Label for Results
result_label = tk.Label(root, text="Detected: ", font=("Arial", 14))
result_label.pack()

# Live Detection Button
live_btn = tk.Button(root, text="Live Webcam Detection", command=live_detection, font=("Arial", 12))
live_btn.pack()

# Run GUI
root.mainloop()
```
